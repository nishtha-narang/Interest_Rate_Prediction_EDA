# Importing the libraries

```python
# Import the necessary/relevant libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings ('ignore')
```

# Loading the datasets

```python
#Load the train data file and check the first 5 records
df1 = pd.read_csv ('loan_data_train.csv')
df1.head()
```

```python
#Check for rows*columns in the data
df1.shape
```

```python
#Load the test data file and check the first 5 records
df2 = pd.read_csv ('loan_data_test.csv')
df2.head()
```

```python
#Check for rows*columns in the data
df2.shape
```

```python
# Add 'Interest.Rate' column in Test Data for easy concatenation and check for first 5 records
df2['Interest.Rate'] = 'testdata'
df2.head()
```

```python
#Concatenate both the files for easy preprocessing
df = pd.concat((df1, df2))
df.head()
```

```python
#Check for concatenated data shape
df.shape
```

```python
#Replace the . in column names with _ for easy accessibility
df.columns = df.columns.str.replace ('.', '_')
df.columns
```

```python
#Segregate the true numeric and true object columns
df.info()
```

```python
#df['Amount_Requested'].astype(float)
df['Amount_Requested'] = df['Amount_Requested'].replace('.', np.nan).astype(float)
```

```python
#df['Amount_Funded_By_Investors'].astype(float)
df.loc[df['Amount_Funded_By_Investors']=='.', 'Amount_Funded_By_Investors'] = np.nan
df['Amount_Funded_By_Investors'] = pd.to_numeric(df['Amount_Funded_By_Investors'], errors='coerce')
```

```python
df['Interest_Rate'].replace('%', '', regex=True, inplace=True)
```

```python
df['Loan_Length'].value_counts()
```

```python
#df['Loan_Length'].astype(float)
df['Loan_Length'] = df['Loan_Length'].replace(' months', '', regex=True)
df['Loan_Length'] = df['Loan_Length'].replace('.', np.nan)
df['Loan_Length'] = df['Loan_Length'].astype(float)
```

```python
#df['Debt_To_Income_Ratio'].astype(float)
df['Debt_To_Income_Ratio'] = df['Debt_To_Income_Ratio'].replace('%', '', regex=True).astype(float)
```

```python
#df['FICO_Range'].astype(float)
x = df['FICO_Range'].str.split('-', expand=True)
x.columns = ['Max', 'Min']
x = x.astype(int)
df['FICO_Range'] = (x['Max']+x['Min'])/2
df['FICO_Range']
```

```python
#df['Open_CREDIT_Lines'].astype(float)
df['Open_CREDIT_Lines'] = df['Open_CREDIT_Lines'].replace('.', np.nan).astype(float)
```

```python
#df['Revolving_CREDIT_Balance'].astype(float)
df['Revolving_CREDIT_Balance'] = df['Revolving_CREDIT_Balance'].replace('.', np.nan).astype(float)
```

```python
#df['Employment_Length'].astype(float)
df['Employment_Length'] = df['Employment_Length'].replace(r'[a-z +]', '', regex=True)
df['Employment_Length'] = df['Employment_Length'].replace('<1', '0.5')
df['Employment_Length'] = df['Employment_Length'].replace('.', np.nan).astype(float)
df['Employment_Length'].value_counts()
```

```python
#Look for changed data types after processing
df.dtypes
```

# Data Preprocessing

```python
#Create a copy of data for further processing
data = df.copy()
```

```python
#Look for duplicate values in the dataset
data.duplicated().sum()
```

```python
#Check rows*columns in the new data
data.shape
```

```python
# Drop columns which may not be significant in the process
data.drop(['ID'],axis=1,inplace=True)
```

# Handling missing values

```python
#Look for missing values in the dataset
data.isnull().sum()/len(data)*100
```

```python
#Look for values in each column separately to handle them accordingly
print(data['Amount_Requested'].value_counts())
print('----------------------------------------')
print(data['Amount_Funded_By_Investors'].value_counts())
print('----------------------------------------')
print(data['Loan_Length'].value_counts())
print('----------------------------------------')
print(data['Loan_Purpose'].value_counts())
print('----------------------------------------')
print(data['Debt_To_Income_Ratio'].value_counts())
print('----------------------------------------')
print(data['State'].value_counts())
print('----------------------------------------')
print(data['Home_Ownership'].value_counts())
print('----------------------------------------')
print(data['Monthly_Income'].value_counts())
print('----------------------------------------')
print(data['FICO_Range'].value_counts())
print('----------------------------------------')
print(data['Open_CREDIT_Lines'].value_counts())
print('----------------------------------------')
print(data['Revolving_CREDIT_Balance'].value_counts())
print('----------------------------------------')
print(data['Inquiries_in_the_Last_6_Months'].value_counts())
print('----------------------------------------')
print(data['Employment_Length'].value_counts())
```

```python
#Imputing missing values in numerical columns with 'median' as it is less sensitive towards outliers
data['Amount_Requested'].replace(np.nan,data['Amount_Requested'].median(), inplace=True)
data['Amount_Funded_By_Investors'].replace(np.nan,data['Amount_Funded_By_Investors'].median(), inplace=True)
data['Debt_To_Income_Ratio'].replace(np.nan,data['Debt_To_Income_Ratio'].median(), inplace=True)
data['Monthly_Income'].replace(np.nan,data['Monthly_Income'].median(), inplace=True)
data['FICO_Range'].replace(np.nan,data['FICO_Range'].median(), inplace=True)
data['Open_CREDIT_Lines'].replace(np.nan,data['Open_CREDIT_Lines'].median(), inplace=True)
data['Revolving_CREDIT_Balance'].replace(np.nan,data['Revolving_CREDIT_Balance'].median(), inplace=True)
data['Inquiries_in_the_Last_6_Months'].replace(np.nan,data['Inquiries_in_the_Last_6_Months'].median(), inplace=True)
data['Employment_Length'].replace(np.nan,data['Employment_Length'].median(), inplace=True)
```

```python
#Imputing missing values in the categorical columns with 'mode'
data['Loan_Length'].replace(np.nan,data['Loan_Length'].mode()[0], inplace=True)
data['Loan_Purpose'].replace(np.nan,data['Loan_Purpose'].mode()[0], inplace=True)
data['State'].replace(np.nan,data['State'].mode()[0], inplace=True)
data['Home_Ownership'].replace(np.nan,data['Home_Ownership'].mode()[0], inplace=True)
```

```python
#Look for any missed missing values to be imputed
data.isnull().sum()
```

```python
#Plot bar charts for numerical columns in order to know their distribution
for col in data.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[col], kde=True, color='skyblue')
    plt.title(f'{col}')
    plt.show()
```

```python
#Capping negative amount in 'Amount_Funded_By_Investors' column to 0 as the amount can't be negative and is nearly close to 0
data['Amount_Funded_By_Investors'] = np.where(data['Amount_Funded_By_Investors']<0,0,data['Amount_Funded_By_Investors'])
```

```python
#Using np.log1p() as it safely log-transforms income, reducing the effect of extreme values while keeping everything interpretable and model-friendly.
data['Monthly_Income'] = np.log1p(data['Monthly_Income'])
```

# Encoding

```python
#Look for unique values in Loan_Purpose column
data['Loan_Purpose'].value_counts()
```

```python
#Group top 2 categories as one and the rest as other for ease of encoding
data['Loan_Purpose'] = data['Loan_Purpose'].apply(lambda x: 1 if x in {'debt_consolidation', 'credit_card'} else 0)
```

```python
#Look for the updated column
data['Loan_Purpose'].value_counts()
```

```python
#Use LabelEncoder for 'State' column as it is more feasible
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['State'] = le.fit_transform(data['State'])
```

```python
#Look for unique values in Home_Ownership column
data['Home_Ownership'].value_counts()
```

```python
#Use dummy variable (one-hot coding) technique to convert categorical to numerical
data=pd.get_dummies(data,columns=['Home_Ownership'],dtype=int)
```
